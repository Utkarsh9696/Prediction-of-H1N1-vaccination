```python
df=df.drop(["census_msa","employment","housing_status","sex","marital_status","income_level","race","qualification","age_bracket"],axis=1)
```

```python
df.dtypes
```

```python
x=df.drop(['h1n1_vaccine'],axis=1)
y=df[['h1n1_vaccine']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
x_train
```

```python
model_ada=HistGradientBoostingClassifier()
model_ada.fit(x_train,y_train)
print(model_ada.score(x_train,y_train))
print(model_ada.score(x_test,y_test))
```
